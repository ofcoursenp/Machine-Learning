```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1)
```

```python
mnist
```

```python
x, y = mnist['data'], mnist['target']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = x[:60000], x[60000:], y[:60000], y[60000:]
```

```python
y_train
```

```python
# from sklearn.linear_model import LogisticRegression
# from sklearn.model_selection import train_test_split
# import numpy as np

# # suffle_index = np.random.permutation(56000)

# # suffle_index = suffle_index.tolist()

# X_train, x_test, y_train, y_test = train_test_split(
#     x, y, test_size=0.2, random_state=42)

# # # x_train, y_train = x_train.iloc[suffle_index], y_train.iloc[suffle_index]
# # # x_train

# y_train


y_train = y_train.astype('int64')
y_test = y_test.astype('int64')

# y_train
```

```python
# y_train = [n == 2 for n in y_train]
```

```python
y_train_2 = [n == 2 for n in y_train]
y_test_2 = (y_test == 2)
```

```python
y_train_2
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(tol=0.2)
```

```python
clf.fit(x_train, y_train_2)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
a = cross_val_score(clf, x_test, y_test_2, cv=3, scoring='accuracy')
```

```python
a.mean()
```

## For 3

## Not 2


```python
y_train_not_2 = [n != 2 for n in y_train]
y_test_not_2 = (y_test != 2)
```

```python
clf = LogisticRegression(tol=0.2)
```

```python
clf.fit(x_train, y_train_2)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(clf, x_test, y_test_not_2)
```

```python
from sklearn.model_selection import cross_val_predict
```

```python
a = cross_val_predict(clf, x_test, y_test_2, cv=3, )
```

```python
a
```

## Calculating Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test_2 , a)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
precision_score(y_test_2 , a)
```

```python
recall_score(y_test_2 , a)
```

```python
f1_score(y_test_2 , a)
```

## Precision Recall Curve

```python
from sklearn.metrics import precision_recall_curve

y_scores = cross_val_predict(clf, x_test, y_test_2, cv=3, method='decision_function')
precision, recall, threshold = precision_recall_curve(y_test_2 , y_scores)
```

```python
precision
```

```python
recall
```

## Precision Recall Curve

```python
%matplotlib inline

import matplotlib.pyplot as plt

plt.plot(threshold, precision[:-1] ,"b--", label='precision')
plt.plot(threshold, recall[:-1], "g-", label='recall')
plt.xlabel('Threshold')
plt.legend(loc='upper left')
plt.ylim([0, 1])
plt.show()
```
